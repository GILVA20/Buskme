*   ¿Cuál es el mayor valor de `NO2` registrado? ¿A qué día pertenece? Mayor NO2: 120.2 en 24/01/2017 19:00
*   ¿Cuál es el menor valor de `PM10` registrado?¿A qué día pertenece? Menor PM10: -5.6 en 12/09/2017 23:00
*   ¿Cuál es la mediana del `NO`? ¿Cómo se interpreta? Mediana NO: 7.1 (valor central de la distribución)
*   ¿Cuál es el primer cuartil de `PM2.5`? ¿Qué significa? 1er cuartil PM2.5: 7.0 (25% de lecturas por debajo)

```python
import pandas as pd
air_df = pd.read_csv('LaqnData.csv')
pvt_df = air_df.pivot(
    index='ReadingDateTime',
    columns='Species',
    values='Value'
)

desc = pvt_df.describe()
max_no2      = desc.loc['max',   'NO2']
date_max_no2 = pvt_df['NO2'].idxmax()
min_pm10     = desc.loc['min',   'PM10']
date_min_pm10= pvt_df['PM10'].idxmin()
median_no    = desc.loc['50%',   'NO']
q1_pm25      = desc.loc['25%',   'PM2.5']

print(f"\nMayor NO2: {max_no2} en {date_max_no2}")
print(f"Menor PM10: {min_pm10} en {date_min_pm10}")
print(f"Mediana NO: {median_no} (valor central de la distribución)")
print(f"1er cuartil PM2.5: {q1_pm25} (25% de lecturas por debajo)")
```

10.	Explica cuáles son las diferencias y similitudes entre el par `melt()/pivot()` y el par `stack()/unstack()`. Si tuvieras que elegir una contraparte para `melt()` entre `stack()/unstack()`, ¿cuál elegirías? ¿Por qué?
melt() convierte columnas en filas; pivot() hace lo contrario.
stack() apila columnas dentro del índice; unstack() las lleva desde el índice a columnas.
melt/pivot trabajan en un DataFrame “plano”; stack/unstack requieren un índice jerárquico.
Yo elegiría stack() como contraparte de melt(), porque ambos “empujan” columnas hacia abajo para convertirlas en filas.
